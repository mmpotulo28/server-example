```python
import socket

# Create a socket object using the socket library
server_socket = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
host = socket.gethostname()
port = 12000
server_socket.bind((host, port))

print("Server is waiting for a connection...")

while True:
    message, address = server_socket.recvfrom(2048)
    modified_message = message.decode().upper()
    print("Message converted to uppercase: ")
    server_socket.sendto(modified_message.encode(), address)
    print(message)
    print("Message sent back to client")
```
