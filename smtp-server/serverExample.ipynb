```python
# sourcery skip: use-assigned-variable
import smtplib

# gmail SMTP server details
smtp_server = 'smtp.gmail.com'
port = 587 # For starttls

# gmail account details
email_address = 'mptdevs@gmail.com'
sender_password = 'fosm zryy mseb jjtb'

# Email content
from_address = email_address
to_address = 'mpotulom28@gmail.com'
subject = 'Test Email from Python'
body = 'This is a test email sent from a Python program using Gmail!'

# Create email message
message = f"""From: {from_address}
To: {to_address}
Subject: {subject}
{body}
"""
server = smtplib.SMTP_SSL(smtp_server, port, timeout=220)

try:
    # Create SMTP session for sending the mail
    
    server.ehlo() # Can be omitted
    server.starttls() # Secure the connection
    server.login(email_address, sender_password)
    # Send the email
    server.sendmail(from_address, to_address, message)
    print("Email sent successfully!")
except Exception as e:
    print(f"Failed to send email: {e}")
finally:
    server.quit()
```
