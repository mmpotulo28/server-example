```python
import socket

# Create a socket object using the socket library
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

# Get the local machine name (hostname)
host = socket.gethostname()

# Set a port for the server that is not in use
port = 12345

# Bind the socket to the host and port
server_socket.bind((host, port))

# Listen for incoming connections (max 5 clients)
server_socket.listen(5)

print("Server is waiting for a connection...")

# Accept any incoming connection
client_socket, addr = server_socket.accept()

print(f"Got a connection from {str(addr)}")

# Receive the message from the client
message = client_socket.recv(1024).decode()

# Check the message and respond
if message == "hello NSS370S class":
    print("Message received:", message)
    response = "Hi, client! Message received."
else:
    response = "Unknown message"

# Send a response back to the client
client_socket.send(response.encode())

# Close the client socket
client_socket.close()

# Close the server socket
server_socket.close()
```
